```python
import pandas as pd
ufo = pd.read_csv('http://bit.ly/uforeports')
ufo.head()
```

```python
#loc filtering or selecting things by labels (rows, column names), : means all columns
ufo.loc[0,:]
```

```python
ufo.loc[[0,1,2], :]
```

```python
#inclusive on the row on both sides
ufo.loc[0:2, :]
```

```python
#default is all inclusive on columns
ufo.loc[0:2]
```

```python
#column selection
ufo.loc[:, ['City','State']].tail()
```

```python
#column selection, city through state, which is a range of columns
ufo.loc[:, 'City':'State'].tail()
```

```python
#alternative way, dropped the Time colunn
ufo.head(3).drop('Time', axis=1)
```

```python
ufo[ufo.City=='Ithaca']
```

```python
#more flexible query, so that you can select your column
ufo.loc[ufo.City=='Oakland',:]
```

```python
#talk about iloc now, select by integer position, position of hte column in dataframe
ufo.iloc[:, [0,3]]
```

```python
#iloc commnand is inclusive of the first number, exclusive of the last number
ufo.iloc[:, 0:4].head()
```

```python
ufo.iloc[0:3,:]
```

```python
#inner bracket is list, passed on to the outer bracket
ufo[['City','State']]
```

```python
ufo.loc[:,['City','State']].tail(2)
```

```python
ufo[0:2]
```

```python
#ix mixes labels and integers, a blend between loc and iloc
```

```python
drinks = pd.read_csv('http://bit.ly/drinksbycountry', index_col='country')
drinks.head()
```

```python
#o below is the integer position of the column; ix will guess using label or integer
drinks.ix['Albania', 0]
```

```python
drinks.ix[1, 'beer_servings']
```

```python
drinks.ix['Albania':'Andorra', 0:2]
```

```python
#integer index will use lable (inclusive on both sides); string index on column will use integeter (not including on the right side)
ufo.ix[0:2,0:2]
```
